# Normalizasyon işlemleri

$\widehat{x_i}=\frac{x_i-x_{min}}{x_{max}-x_{min}}$

```python
import pandas as pd
import numpy as np
```

```python
veri_seti = pd.read_csv("titanic.csv")
```

```python
veri_seti.head()
```

```python
veri_seti.Age.fillna(veri_seti.Age.mean(), inplace=True)
```

```python
# Min-max normalizasyonu
mm_normalize_edildi = (veri_seti.Age - veri_seti.Age.min()) / (
    veri_seti.Age.max() - veri_seti.Age.min())
```

```python
veri_seti.Age.min()
```

```python
veri_seti.Age.max()
```

```python
mm_normalize_edildi.head()
```

Standardizasyon
z-skor normalizasyonu olarak da bilinir. Bu metotla biz verilerin dağılımını ortalaması 0 standart sapması 1 olan bir dağılıma uydurmaya çalışıyoruz. Standardizasyon denmesinin nedeni budur. Standart normal dağılımdan dolayı geliyor. Standart normal dağılımı hatırlıyor musunuz? Nedir?

```python
# standardizasyon
standardize_edildi = (veri_seti.Age -
                      veri_seti.Age.mean()) / veri_seti.Age.std()
```
